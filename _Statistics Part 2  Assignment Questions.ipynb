```python
# Q2.Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).
# Ans=
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

# Define the range of the discrete uniform distribution
low = 1
high = 7
x = np.arange(low, high + 1)

# Calculate the PMF
pmf = st.randint.pmf(x, low, high)

# Plot the PMF
plt.bar(x, pmf, color='skyblue')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Discrete Uniform Distribution PMF')
plt.xticks(x)
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# Q3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
# Ans=
def bernoulli_pmf(x, p):
    """
    Calculate the PMF of a Bernoulli distribution.

    Parameters:
    x : int
        The value (0 or 1) at which to evaluate the PMF.
    p : float
        Probability of success (value 1), 0 <= p <= 1.

    Returns:
    float
        PMF value at x.
    """
    if x not in [0, 1]:
        return 0  # Bernoulli is only defined for 0 or 1
    
    return p**x * (1 - p)**(1 - x)

# Example usage:
p = 0.7
print("P(X=1):", bernoulli_pmf(1, p))
print("P(X=0):", bernoulli_pmf(0, p))
```

```python
# Q6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete 
# uniform distribution
# Ans=
import numpy as np
import matplotlib.pyplot as plt

# Define the discrete uniform distribution range
start = 1
end = 6
values = np.arange(start, end + 1)

# Probability for each value (equal)
probability = 1 / len(values)

# Calculate the PMF (all equal)
pmf = np.full_like(values, probability, dtype=float)

# Calculate the CDF by cumulative sum of PMF
cdf = np.cumsum(pmf)

# Plot the CDF as a step plot
plt.step(values, cdf, where='post', label='CDF', color='blue')
plt.scatter(values, cdf, color='red')  # Points on the CDF steps

plt.title('CDF of Discrete Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.xticks(values)
plt.ylim(0, 1.05)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Q9. Write a Python function to calculate Z-scores from a dataset and plot them
# Ans=
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

def plot_zscores(data):
    """
    Calculates Z-scores for a dataset and plots them.

    Args:
      data (list or numpy.ndarray): The input dataset.

    Returns:
      None. Displays the plot of Z-scores.
    """
    z_scores = zscore(data)
    
    plt.figure(figsize=(10, 6))
    plt.plot(z_scores, marker='o', linestyle='-')
    plt.title('Z-Scores Plot')
    plt.xlabel('Data Point Index')
    plt.ylabel('Z-Score')
    plt.grid(True)
    plt.axhline(0, color='black', linestyle='--')
    plt.show()

if __name__ == '__main__':
    # Example usage:
    example_data = np.array([10, 12, 15, 13, 18, 20, 22, 19, 25, 28, 30, 35, 12, 14, 16])
    plot_zscores(example_data)

    another_data = [1, 2, 2, 2, 3, 1, 1, 15, 2, 2, 2, 3, 1, 1, 2]
    plot_zscores(another_data)
```

```python
# Q13.Generate random variables and calculate their corresponding probabilities using the binomial distribution
# Ans=
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters for binomial distribution
n = 10       # number of trials
p = 0.5      # probability of success
size = 1000  # number of random variables to generate

# Step 1: Generate random variables
random_vars = np.random.binomial(n, p, size)

# Step 2: Calculate PMF for all possible outcomes
x = np.arange(0, n + 1)        # possible number of successes
pmf = binom.pmf(x, n, p)       # theoretical PMF

# Step 3: Count occurrences of each outcome in the sample
values, counts = np.unique(random_vars, return_counts=True)
relative_freq = counts / size  # approximate empirical probability

# Step 4: Display probabilities
print("Theoretical PMF:")
for xi, prob in zip(x, pmf):
    print(f"P(X = {xi}) = {prob:.4f}")

print("\nEmpirical Probabilities (from simulation):")
for v, f in zip(values, relative_freq):
    print(f"P(X = {v}) â‰ˆ {f:.4f}")

# Step 5: Plot both PMF and empirical distribution
plt.bar(x, pmf, width=0.4, label='Theoretical PMF', align='edge', color='skyblue', edgecolor='black')
plt.bar(values - 0.4, relative_freq, width=0.4, label='Empirical Frequencies', align='edge', color='orange', edgecolor='black')

plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.xticks(x)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Q14.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal 
# distribution
# Ans=
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std_dev):
    """
    Calculate the Z-score for a data point x.
    """
    return (x - mean) / std_dev

# Example inputs
x_value = 75          # Data point
population_mean = 70  # Mean of the distribution
std_dev = 10          # Standard deviation

# Step 1: Calculate Z-score
z_score = calculate_z_score(x_value, population_mean, std_dev)
print(f"Z-score for x = {x_value} is: {z_score:.2f}")

# Step 2: Plot standard normal distribution
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, 0, 1)

plt.plot(x, pdf, label='Standard Normal Distribution', color='blue')
plt.fill_between(x, pdf, where=(x < z_score), color='skyblue', alpha=0.5)

# Mark the Z-score
plt.axvline(z_score, color='red', linestyle='--', label=f'Z = {z_score:.2f}')
plt.title('Z-score Comparison to Standard Normal Distribution')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Q15.Implement hypothesis testing using Z-statistics for a sample dataset
# Ans=
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [102, 100, 98, 97, 105, 101, 99, 103, 100, 98]
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Population mean (known or assumed)
population_mean = 100

# Significance level
alpha = 0.05

# Step 1: Calculate Z-statistic
z = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-statistic = {z:.4f}")

# Step 2: Find critical Z-value for two-tailed test
z_critical = norm.ppf(1 - alpha / 2)
print(f"Critical Z-value = Â±{z_critical:.4f}")

# Step 3: Decision
if abs(z) > z_critical:
    print("Reject the null hypothesis (Hâ‚€): Sample mean is significantly different.")
else:
    print("Fail to reject the null hypothesis (Hâ‚€): No significant difference.")
```

```python
# Q16. Create a confidence interval for a dataset using Python and interpret the result
# Ans=
import numpy as np
from scipy.stats import t

# Sample dataset
data = [52, 55, 60, 63, 58, 57, 54, 59, 61, 62]

# Step 1: Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation
n = len(data)

# Step 2: Set confidence level
confidence = 0.95
alpha = 1 - confidence

# Step 3: Calculate t-critical value (since population std is unknown)
t_critical = t.ppf(1 - alpha/2, df=n-1)

# Step 4: Calculate margin of error
margin_of_error = t_critical * (sample_std / np.sqrt(n))

# Step 5: Calculate confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Output results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"{int(confidence*100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
# Q17.Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
# Ans=
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt

# Step 1: Generate data from a normal distribution
np.random.seed(0)  # for reproducibility
mu, sigma = 50, 10  # true population mean and std dev
n = 30              # sample size

sample = np.random.normal(mu, sigma, n)

# Step 2: Calculate sample statistics
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)

# Step 3: Confidence level
confidence = 0.95
alpha = 1 - confidence
df = n - 1

# Step 4: Calculate the t-critical value
t_crit = t.ppf(1 - alpha/2, df)

# Step 5: Margin of error
margin_of_error = t_crit * (sample_std / np.sqrt(n))

# Step 6: Confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"{int(confidence*100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

# Optional: Plot the sample and CI
plt.hist(sample, bins=10, color='lightblue', edgecolor='black')
plt.axvline(sample_mean, color='red', linestyle='--', label=f'Mean = {sample_mean:.2f}')
plt.axvline(lower_bound, color='green', linestyle='--', label='Confidence Interval')
plt.axvline(upper_bound, color='green', linestyle='--')
plt.title('Histogram of Sample with 95% Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Q1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and 
# interpret the results
# Ans=
import numpy as np
from scipy.stats import norm

# Step 1: Simulate sample data
np.random.seed(0)
sample = np.random.normal(loc=102, scale=15, size=50)  # Sample from N(102, 15)

# Step 2: Known population parameters
population_mean = 100
population_std = 15  # Known population std

# Step 3: Compute sample statistics
sample_mean = np.mean(sample)
sample_size = len(sample)

# Step 4: Compute Z-statistic
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Step 5: Calculate two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Step 6: Output
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Score: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")

# Step 7: Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Result: Reject the null hypothesis â€” significant difference from population mean.")
else:
    print("Result: Fail to reject the null hypothesis â€” no significant difference.")
```

```python
# Q4.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
# Ans=
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu_0=100, mu_1=105, sigma=10, n=30, alpha=0.05):
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value for Type I error (right-tailed test)
    z_alpha = norm.ppf(1 - alpha)
    x_crit = mu_0 + z_alpha * se

    # Type II error: probability of accepting H0 when H1 is true
    beta = norm.cdf((x_crit - mu_1) / se)

    # Visualization setup
    x = np.linspace(mu_0 - 4*se, mu_1 + 4*se, 1000)
    y_H0 = norm.pdf(x, mu_0, se)
    y_H1 = norm.pdf(x, mu_1, se)

    plt.figure(figsize=(10, 6))

    # Plot both distributions
    plt.plot(x, y_H0, label="Null Hypothesis $H_0$", color="blue")
    plt.plot(x, y_H1, label="Alternative Hypothesis $H_1$", color="green")

    # Shade Type I error (Î±): area in H0 beyond critical value
    x_alpha = x[x > x_crit]
    plt.fill_between(x_alpha, norm.pdf(x_alpha, mu_0, se), color='red', alpha=0.5, label='Type I Error (Î±)')

    # Shade Type II error (Î²): area in H1 before critical value
    x_beta = x[x < x_crit]
    plt.fill_between(x_beta, norm.pdf(x_beta, mu_1, se), color='orange', alpha=0.5, label='Type II Error (Î²)')

    # Plot critical value
    plt.axvline(x_crit, color='black', linestyle='--', label=f'Critical Value = {x_crit:.2f}')

    plt.title("Type I and Type II Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print numerical results
    print(f"Critical Value: {x_crit:.2f}")
    print(f"Type I Error (Î±): {alpha:.4f}")
    print(f"Type II Error (Î²): {beta:.4f}")
    print(f"Power of the test (1 - Î²): {1 - beta:.4f}")

# Example usage
visualize_type1_type2_errors(mu_0=100, mu_1=105, sigma=10, n=30, alpha=0.05)
```

```python
# Q7 Perform a two-tailed Z-test using Python and visualize the decision region on a plot
# Ans=
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05):
    # Calculate standard error
    se = population_std / np.sqrt(sample_size)
    
    # Calculate Z-score
    z_score = (sample_mean - population_mean) / se
    
    # Critical values for two-tailed test
    z_critical = norm.ppf(1 - alpha/2)
    
    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))
    
    print(f"Z-Score: {z_score:.4f}")
    print(f"P-Value: {p_value:.4f}")
    
    # Plot standard normal distribution
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)
    plt.plot(x, y, label='Standard Normal Distribution')
    
    # Shade rejection regions
    x_reject_left = np.linspace(-4, -z_critical, 200)
    x_reject_right = np.linspace(z_critical, 4, 200)
    plt.fill_between(x_reject_left, norm.pdf(x_reject_left), color='red', alpha=0.5, label='Rejection Region')
    plt.fill_between(x_reject_right, norm.pdf(x_reject_right), color='red', alpha=0.5)
    
    # Plot observed z_score
    plt.axvline(z_score, color='blue', linestyle='--', label=f'Observed Z = {z_score:.2f}')
    
    # Plot critical values
    plt.axvline(z_critical, color='black', linestyle='--', label=f'Critical Z = Â±{z_critical:.2f}')
    plt.axvline(-z_critical, color='black', linestyle='--')
    
    plt.title('Two-Tailed Z-Test')
    plt.xlabel('Z-value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
population_mean = 100
population_std = 15
sample_size = 40
sample_mean = 105  # Sample mean to test
alpha = 0.05

two_tailed_z_test(sample_mean, population_mean, population_std, sample_size, alpha)
```

```python
# Q9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance
# Ans=
import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence=0.95):
    """
    Calculate the confidence interval for the sample mean.
    
    Parameters:
    - sample: array-like, the data sample
    - confidence: float, confidence level (default 0.95 for 95%)
    
    Returns:
    - (lower_bound, upper_bound): tuple of floats representing the confidence interval
    """
    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)  # sample standard error
    
    # Use t-distribution critical value for confidence interval
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)
    
    margin_of_error = t_crit * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error
    
    return lower_bound, upper_bound

# Example usage:
np.random.seed(0)
sample_data = np.random.normal(loc=100, scale=15, size=40)

ci_lower, ci_upper = confidence_interval(sample_data, confidence=0.95)
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
# Q10.Write a Python program to calculate the margin of error for a given confidence level using sample data
# Ans=
import numpy as np
from scipy.stats import t

def calculate_margin_of_error(sample, confidence=0.95):
    """
    Calculate the margin of error for the sample mean at the specified confidence level.
    
    Parameters:
    - sample: array-like, sample data
    - confidence: float, confidence level (default 0.95)
    
    Returns:
    - margin_of_error: float
    """
    n = len(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)  # sample standard error
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)  # critical t value
    margin_of_error = t_crit * std_err
    return margin_of_error

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=8, size=30)
me = calculate_margin_of_error(sample_data, confidence=0.95)
print(f"Margin of Error at 95% confidence: {me:.3f}")
```

```python
# Q11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
# Ans=
def bayes_theorem(prior, likelihood, marginal_likelihood):
    """
    Calculate posterior probability using Bayes' theorem.
    
    prior: P(H) - prior probability of hypothesis
    likelihood: P(E|H) - likelihood of evidence given hypothesis
    marginal_likelihood: P(E) - total probability of evidence
    
    Returns posterior probability P(H|E)
    """
    return (likelihood * prior) / marginal_likelihood

# Given data
P_disease = 0.01            # Prior probability of disease
P_no_disease = 1 - P_disease

P_pos_given_disease = 0.99  # Sensitivity
P_neg_given_no_disease = 0.95  # Specificity
P_pos_given_no_disease = 1 - P_neg_given_no_disease

# Marginal likelihood P(Pos)
P_pos = (P_pos_given_disease * P_disease) + (P_pos_given_no_disease * P_no_disease)

# Calculate posterior probability P(Disease|Pos)
posterior = bayes_theorem(P_disease, P_pos_given_disease, P_pos)

print(f"Probability of disease given a positive test: {posterior:.4f} ({posterior*100:.2f}%)")
```

```python
# Q13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed 
# data
# Ans=
import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for Chi-square test given observed frequencies.
    
    Parameters:
    - observed: 2D numpy array of observed frequencies
    
    Returns:
    - expected: 2D numpy array of expected frequencies
    """
    row_sums = observed.sum(axis=1, keepdims=True)
    col_sums = observed.sum(axis=0, keepdims=True)
    total = observed.sum()
    
    expected = (row_sums @ col_sums) / total
    return expected

# Example observed data (contingency table)
observed = np.array([
    [20, 30, 25],
    [10, 20, 15],
    [30, 10, 20]
])

expected = calculate_expected_frequencies(observed)

print("Observed Frequencies:\n", observed)
print("\nExpected Frequencies:\n", expected)
```
